```python
%matplotlib inline 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import requests
import StringIO
import zipfile
import scipy.stats
import seaborn as sns
import sklearn
import sklearn.cross_validation
```

First the pickle file is read in. This includes all columns of interest that were pulled. Note that the only entries that were pulled were those with more than 0 undergrad checkout score.

```python
df = pd.read_pickle('C:/Users/mshand/Downloads/Stuff.pkl')
```

There are no 0 checkout scores recorded, so NA values are in fact the same as 0 values in this case. All graduate and faculty checkout scores must have the NA values replaced with 0s.

```python
df.score_checkouts_grad.fillna(0, inplace=True)
df.score_checkouts_fac.fillna(0, inplace=True)
```

Also, there were less than 10 entries with NA values for undergraduate checkout score. These appeared to be NA values across all rows and it made the most sense to simply drop them from the dataframe.

```python
df.dropna(subset=['score_checkouts_undergrad'], inplace=True)
```

In order to create some graphs that are in expensive in computation or memory we broke the data into a random sample a third of the original size. This allowed easier use for certain graphs whose overall pattern should not be effected by taking a sample.

```python
sampleLarge, sampleSmall = sklearn.cross_validation.train_test_split(df, test_size=.33, random_state=42)
```

Then the smaller sample needed to be restored as a dataframe with the correct columns.

```python
smallDF = pd.DataFrame(sampleSmall, columns=df.columns)
```

We were first interested in what kind of distribution these checkout scores could be modeled with, so we wanted to see the histograms of each group's scores. We restricted the range to better show the pattern for smaller checkout scores, where the majority of entries lie in the first place. Outliers above a score of 50 were therefore excluded from the graph.

```python
plt.hist(smallDF.score_checkouts_undergrad, range=(0,50), bins=50)
plt.title('Undergraduate Checkout Scores')
plt.xlabel('Undergraduate Score')
plt.ylabel('Frequency')
plt.show()

plt.hist(smallDF.score_checkouts_grad, range=(0,50), bins=50)
plt.title('Graduate Checkout Scores')
plt.xlabel('Graduate Score')
plt.ylabel('Frequency')
plt.show()

plt.hist(smallDF.score_checkouts_fac, range=(0,50), bins=50)
plt.title('Faculty Checkout Scores')
plt.xlabel('Faculty Score')
plt.ylabel('Frequency')
plt.show()
```

Undergraduate checkout scores exclude the 0 entries, but we can safely assume from looking at the original data that the number of entries with an undergraduate checkout score of 0 matches the same pattern as the Graduate and Faculty scores. A good fit for this data might be the Zero Inflated Poisson, because the pattern after the 0 entries appears to fit a Poisson (as this is a discrete case), except for the high number of 0 score entries.

We also wanted to compare checkout scores by format.

```python
filterdf = smallDF[['score_checkouts_undergrad', 'score_checkouts_grad', 'score_checkouts_fac', 'format']]

filterdf.score_checkouts_undergrad = np.log(filterdf.score_checkouts_undergrad.astype(float))
sns.boxplot(filterdf.score_checkouts_undergrad, groupby=filterdf.format)
sns.set_context('poster')
plt.title('Undergraduate Scores by Format')
plt.xticks(rotation = 90)
plt.ylabel('Log of Undergraduate Score')
plt.xlabel('')
plt.show()

filterdf.score_checkouts_grad = np.log(filterdf.score_checkouts_grad.astype(float))
sns.boxplot(filterdf.score_checkouts_grad, groupby=filterdf.format)
sns.set_context('poster')
plt.title('Graduate Scores by Format')
plt.xticks(rotation = 90)
plt.ylabel('Graduate Score')
plt.xlabel('')
plt.show()

filterdf.score_checkouts_fac = np.log(filterdf.score_checkouts_fac.astype(float))
sns.boxplot(filterdf.score_checkouts_fac, groupby=filterdf.format)
sns.set_context('poster')
plt.title('Faculty Scores by Format')
plt.xticks(rotation = 90)
plt.ylabel('Faculty Score')
plt.xlabel('')
plt.show()
```

While the boxplots certainly have more information than a bar graph, perhpas a bar graph with all three groups on the same chart could give a more succinct view.

```python
filterdf = smallDF[['score_checkouts_undergrad', 'score_checkouts_grad', 'score_checkouts_fac', 'format']]
filterdf.score_checkouts_undergrad = filterdf.score_checkouts_undergrad.astype(float)
filterdf.score_checkouts_grad = filterdf.score_checkouts_grad.astype(float)
filterdf.score_checkouts_fac = filterdf.score_checkouts_fac.astype(float)
grouped = filterdf.groupby('format')
means = grouped.mean()

means = means.reindex()
newMeans = pd.DataFrame(index = range((len(means)-1)*3), columns = ('Format','Score','Group'))

for i in range(len(means)-1):
    j = i*3
    newMeans['Format'][j] = means.index[i]
    newMeans['Format'][j+1] = means.index[i]
    newMeans['Format'][j+2] = means.index[i]
    newMeans.Score[j] = means.score_checkouts_undergrad[i]
    newMeans.Group[j] = 'Undergrad'
    newMeans.Score[j+1] = means.score_checkouts_grad[i]
    newMeans.Group[j+1] = 'Grad'
    newMeans.Score[j+2] = means.score_checkouts_fac[i]
    newMeans.Group[j+2] = 'Faculty'
    
sns.factorplot('Format', 'Score', 'Group', newMeans, kind='bar', palette="Blues", aspect=1.25)
sns.set_context('poster')
plt.xticks(rotation = 90)
plt.ylabel('Mean Score')
plt.show()
```

Now what was the difference for holding libraries, which could also possibly shed some insight on which schools or subjects have the highest readership.

```python
holdingDF = smallDF[['id','score_checkouts_undergrad','score_checkouts_grad','score_checkouts_fac','holding_libs']].dropna(subset=['holding_libs'])
holdingDF = holdingDF.reindex(index=range(len(holdingDF)))
```

```python
newDF = pd.DataFrame(columns = ('id','score_checkouts_undergrad','score_checkouts_grad','score_checkouts_fac','holding_lib'), index=range(331635))
summ = 0
for i in range(len(holdingDF)):
    if(type(holdingDF.holding_libs[i])== list):
        for j in range(len(holdingDF.holding_libs[i])):
            newDF.id[summ] = holdingDF.id[i]
            newDF.score_checkouts_undergrad[summ] = holdingDF.score_checkouts_undergrad[i]
            newDF.score_checkouts_grad[summ] = holdingDF.score_checkouts_grad[i]
            newDF.score_checkouts_fac[summ] = holdingDF.score_checkouts_fac[i]
            newDF.holding_lib[summ] = holdingDF.holding_libs[i][j]
            
            summ = summ+1
        
newDF.head()
```

```python
logDF = pd.DataFrame(columns = newDF.columns)

logDF.score_checkouts_undergrad = np.log(newDF.score_checkouts_undergrad.astype(float))
sns.set_context({"figure.figsize": (17, 6)})
sns.boxplot(logDF.score_checkouts_undergrad, groupby=newDF.holding_lib)
sns.set_context('poster')
plt.title('Undergraduate Scores by Holding Library')
plt.xticks(rotation = 90)
plt.ylabel('Undergraduate Score')
plt.xlabel('')
plt.show()

logDF.score_checkouts_grad = np.log(newDF.score_checkouts_grad.astype(float))
sns.set_context({"figure.figsize": (17, 6)})
sns.boxplot(logDF.score_checkouts_grad, groupby=newDF.holding_lib)
sns.set_context('poster')
plt.title('Graduate Scores by Holding Library')
plt.xticks(rotation = 90)
plt.ylabel('Graduate Score')
plt.xlabel('')
plt.show()

logDF.score_checkouts_fac = np.log(newDF.score_checkouts_fac.astype(float))
sns.set_context({"figure.figsize": (17, 6)})
sns.boxplot(logDF.score_checkouts_fac, groupby=newDF.holding_lib)
sns.set_context('poster')
plt.title('Faculty Scores by Holding Library')
plt.xticks(rotation = 90)
plt.ylabel('Faculty Score')
plt.xlabel('')
plt.show()
```
